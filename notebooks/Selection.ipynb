```python
import pandas as pd
```

```python
# find . -iname '*.pdf' -exec pdfgrep -i "deep learning" {} + | awk '{ split($0,a,":"); print a[1]  }' | uniq | awk '{ split($0,a,"_"); printf "%s %s\n",a[1],a[2]  }' | uniq > deeplearning.txt

def readFile(column, filename):
    students_py = pd.DataFrame(columns=[column])
    file_python = open(filename, 'r')
    for name in file_python:
        #print name[3:]    
        students_py.loc[name[2:].split('\n')[0]] = [True]
    return students_py


allfiles = []

allfiles.append(readFile('python', '/Users/cerminar/Downloads/SUMM2018/python.txt'))
allfiles.append(readFile('ml', '/Users/cerminar/Downloads/SUMM2018/machinelearning.txt'))
allfiles.append(readFile('keras', '/Users/cerminar/Downloads/SUMM2018/keras.txt'))

allfiles.append(readFile('dl', '/Users/cerminar/Downloads/SUMM2018/deeplearning.txt'))
allfiles.append(readFile('datascience', '/Users/cerminar/Downloads/SUMM2018/datascience.txt'))

allfiles.append(readFile('tensorflow', '/Users/cerminar/Downloads/SUMM2018/tensorflow.txt'))

#allfiles.append(readFile('jupiter', '/Users/cerminar/Downloads/SUMM2018/jupiter.txt'))

allfiles.append(readFile('anomaly', '/Users/cerminar/Downloads/SUMM2018/anomaly.txt'))


students = pd.concat(allfiles, axis=1)   
    
students
```
