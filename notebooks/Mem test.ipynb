```python
file_default = open(r"/Users/cerminar/Downloads/mem1.out","r")

line = file_default.readline()
cnt = 1
memory_rss = []
memory_vsize = []

events = []
while line:
#     print("Line {}: {}".format(cnt, line.strip()))
    line = file_default.readline()
    if 'RSS' in line:
        cnt += 1
        if cnt%10 == 0 or True:
            print("Line {}: {}".format(cnt, line.strip()))
            events.append(cnt)
            rss = float(line.split()[7])
            vsize = float(line.split()[4])

            print 'RSS: {}'.format(rss)
            memory_rss.append(rss)
            memory_vsize.append(vsize)
            
file_default.close()            
```

```python
file_hyb = open(r"/Users/cerminar/Downloads/mem2.out","r")

line = file_hyb.readline()
cnt = 1
hyb_memory_rss = []
hyb_memory_vsize = []

hyb_events = []
while line:
#     print("Line {}: {}".format(cnt, line.strip()))
    line = file_hyb.readline()
    if 'RSS' in line:
        cnt += 1
        if cnt%10 == 0 or True:
            print("Line {}: {}".format(cnt, line.strip()))
            hyb_events.append(cnt)
            rss = float(line.split()[7])
            vsize = float(line.split()[4])

            print 'RSS: {}'.format(rss)
            hyb_memory_rss.append(rss)
            hyb_memory_vsize.append(vsize)
            
file_hyb.close()            
```

```python
import ROOT
ROOT.enableJSVis()

import array
graph_rss = ROOT.TGraph(len(events), array.array('d', events), array.array('d', memory_rss))
graph_rss.SetMarkerStyle(7)
graph_rss.SetMarkerColor(1)

graph_vsize = ROOT.TGraph(len(events), array.array('d', events), array.array('d', memory_vsize))
graph_vsize.SetMarkerStyle(7)
graph_vsize.SetMarkerColor(2)
```

```python
hyb_graph_rss = ROOT.TGraph(len(hyb_events), array.array('d', hyb_events), array.array('d', hyb_memory_rss))
hyb_graph_rss.SetMarkerStyle(24)
hyb_graph_rss.SetMarkerColor(4)

hyb_graph_vsize = ROOT.TGraph(len(hyb_events), array.array('d', hyb_events), array.array('d', hyb_memory_vsize))
hyb_graph_vsize.SetMarkerStyle(24)
hyb_graph_vsize.SetMarkerColor(5)
```

```python
c = ROOT.TCanvas()
leg = ROOT.TLegend()

graph_vsize.Draw('AP')
leg.AddEntry(graph_vsize, 'VSIZE no hybrid', 'p')
# c.cd()
graph_rss.Draw('P,same')
leg.AddEntry(graph_rss, 'RSS no hybrid', 'p')

hyb_graph_rss.Draw('P,same')
leg.AddEntry(hyb_graph_rss, 'RSS hybrid', 'p')

hyb_graph_vsize.Draw('P,same')
leg.AddEntry(hyb_graph_vsize, 'VSIZE hybrid', 'p')


c.Update()

graph_vsize.GetYaxis().SetLimits(0., 3000.)
graph_vsize.GetYaxis().SetRangeUser(0., 3000.)

graph_rss.GetYaxis().SetLimits(0., 3000.)
leg.Draw('same')
c.Draw()

```
